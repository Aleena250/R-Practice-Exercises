# Lists
Lists are the R objects which contain elements of different types like âˆ’ numbers, strings, vectors and another list inside it. A list can also contain a matrix or a function as its elements.

## Creating List
List is created using list("Enter Elements of list separated by coma") function.

## Naming List Elements
The list elements can be given names and they can be accessed using these names("Enter Name of List Object").

## Accessing List Elements
Elements of the list can be accessed by the index of the element in the list. we can use "[]" to access the index with List Object.
In case of named lists it can also be accessed using the names.

## Manipulating List Elements
We can add, delete and update list elements as shown below. We can add elements only at the end of a list. Because if we add element in between the List, we are ending up updating that particular element and missing the data of that particular index. Hence we can update any element but add element at the end of the list only. But we can delete any element of List.

## Merging Lists
You can merge many lists into one list by placing all the lists inside  c() function.

## Converting List to Vector
A list can be converted to a vector so that the elements of the vector can be used for further manipulation. All the arithmetic operations on vectors can be applied after the list is converted into vectors. To do this conversion, we use the unlist() function. It takes the list as input and produces a vector.



```r
###Run this code......
# Create a list containing a vector, a matrix and a list.
list_data <- list(
                 c("Jan","Feb","Mar"), # character vector
                  matrix(c(3,9,5,1,-2,8), nrow = 2), #matrix
                  list("green",12.3) #inner list
                 )
   
## show list_data element
print(list_data)
```

```r
# Naming List "list_data" Elements 

# Give names to the elements in the list.
names(list_data) <- c("1st Quarter", "A_Matrix", "A Inner list")

# Show the list.
print(list_data)
```

```r
 ## Accessing List Elements using indexed notation

# Access the first element of the list.
print(list_data[1])

# Access the list element using the name of the element.

#Error in code line
#Comment out this line and then run
print(list_data$1st Quarter)

#if our element name starts with digit,we have to  convert our element name as character first and then access it.
#this error occurs because the element name starts with a digit.
print(list_data$'1st Quarter')
```

```r
## Manipulating List Elements

# Add / Update element in the list.
print("Original list before adding any element")
print(list_data)
# Add element at the end of the list.
list_data[4] <- "New element"

print("Updated list after adding element at the end")
print(list_data)

#Now attempt to add element in between List elements,
#we are ending up updating 3rd element and lost the previous value of 3rd element.
list_data[3] <- "New Element 2"

print("Updated list after adding element in between the List Elements")
print(list_data)

```

```r
## Manipulating List Elements

# Delete the element of List 
print("Original list before deleting any element")
print(list_data)
print(paste("Length of List before deleting any element: ", length(list_data)))
# Remove the last element.
list_data[4] <- NULL

print("Updated list after deleting last element")
print(list_data)
print(paste("Length of List after deleting last element: ", length(list_data)))

#Remove the element  in between the list elements
list_data[2] <- NULL

print("Updated list after deleting any element")
print(list_data)
print(paste("Length of List after deleting any element: ", length(list_data)))
```

```r
## Merging Lists
# Create two lists.

list1 <- list(1,2,3)
list2 <- list("Sun","Mon","Tue")

# Merge the two lists using c().
merged.list1 <- c(list1,list2)

# Print the merged list.
print(merged.list1)


```

```r
##Converting List to Vector

# Create lists.
list1 <- list(1:5)
print("List 1")
print(list1)

list2 <-list(10:14)
print("List 2")
print(list2)

# Convert the lists to vectors.
v1 <- unlist(list1)
v2 <- unlist(list2)


print("Vector 1")
print(v1)
print("Vector 2")
print(v2)

# Now add the vectors
result <- v1+v2
print("result ")
print(result)


# Character Type List to Vectors
#We cannot perform arthimatic operations on character list elements
list3 <- list("Red")
print("List 3")
print(list3)
list4 <- list("Blue")
print("List 4")
print(list4)

v3 <- unlist(list3)
v4 <- unlist(list4)

print("Vector 3")
print(v3)
print("Vector 4")
print(v4)

# Un comment this to check the error
#result1 <-v3+v4
#print("result 1")
#print(result1)



# matrix type lists to vector
list5 <- list(matrix(c(3,9,5,1,-2,8), nrow = 2))
print("List 5")
print(list5)
list6 <- list(matrix(c(3,9,5,1,-2,8), nrow = 2))
print("List 6")
print(list6)

v5 <- unlist(list5)
v6 <- unlist(list6)
print("Vector 5")
print(v5)
print("Vector 6")
print(v6)

result2 <- v5+v6
print("result 2")
print(result2)

#Mixed Typed List to Vector (matrix and vector)
list7 <- list(matrix(c(3,9,5,1,-2,8), nrow = 2), #matrix
                  c(3,9,5,1,-2,8) #vector
                 )
print("List 7")
print(list7)

list8 <- list(matrix(c(3,9,5,1,-2,8), nrow = 2), #matrix
                  c(3,9,5,1,-2,8) #vector
                 )
print("List 8")
print(list8)

v7 <- unlist(list7)
v8 <- unlist(list8)
print("Vector 7")
print(v7)
print("Vector 8")
print(v8)


result3 <- v7+v8
print("result 3")
print(result3)
```
